**Importo la biblioteca Pandas**

```python
import pandas as pd
```

**Busco en el repositorio los URL de los raw csv**

```python
url_vehiculos = 'https://raw.githubusercontent.com/woodnathan/MarioKart8-Stats/master/bodies.csv'
```

```python
url_personajes = 'https://raw.githubusercontent.com/woodnathan/MarioKart8-Stats/master/characters.csv'
```

```python
url_neumaticos = 'https://raw.githubusercontent.com/woodnathan/MarioKart8-Stats/master/tires.csv'
```

```python
url_alas = 'https://raw.githubusercontent.com/woodnathan/MarioKart8-Stats/master/gliders.csv'
```

**Obtengo los dataframes de vehiculos, personajes, neumaticos y alas**

```python
vehiculos = pd.read_csv(url_vehiculos, error_bad_lines=False)
```

```python
personajes = pd.read_csv(url_personajes, error_bad_lines=False)
```

```python
neumaticos = pd.read_csv(url_neumaticos, error_bad_lines=False)
```

```python
alas = pd.read_csv(url_alas, error_bad_lines=False)
```

**a) ¿Cuales son los personajes con mejor stat de velocidad?**

```python
personajes[personajes['Speed'] == personajes['Speed'].max()].Character
```

**b) ¿Cuantas posibles combinaciones de personaje-vehiculo-neumaticos-ala se pueden armar?**

```python
cantidadDeCombinacionesPosibles = len(personajes)*len(vehiculos)*len(neumaticos)*len(alas)
print(cantidadDeCombinacionesPosibles)
```
